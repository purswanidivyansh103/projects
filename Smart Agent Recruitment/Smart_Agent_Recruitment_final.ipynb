# Smart Agent Recruitment

## Problem Statement

Smart Agent Recruitment Challenge

FinMan is a Financial Distribution company. Over the last 10 years, they have created an offline distribution channel across India. They sell Financial products to consumers by hiring agents in their network. These agents are freelancers and get commission when they make a product sale.


Overview of FinMan On-boarding process
The Managers at FinMan are primarily responsible for recruiting agents. Once a manager has identified a potential applicant, the would explain the business opportunity to the agent. Once the agent provides the consent, an application is made to FinMan to become an agent. This date is known as application_receipt_date.

In the next 3 months, this potential agent has to undergo a 7 day training at the FinMan branch (about Sales processes and various products) and clear a subsequent examination in order to become a FinMan agent.


The problem - Who are the best agents?
As is obvious in the above process, there is a significant investment which FinMan makes in identifying, training and recruiting these agents. However, there are a set of agents who do not bring in the expected resultant business.

## Feature Engineering and EDA

```python
# Creating Features like Manager_promotion, Pin_difference & Manager_Num_Products_new

df_train['Manager_Promotion']=df_train['Manager_Current_Designation']-df_train['Manager_Joining_Designation']
df_test['Manager_Promotion']=df_test['Manager_Current_Designation']-df_test['Manager_Joining_Designation']

df_train['PinDifference'] =abs(df_train['Office_PIN']-df_train['Applicant_City_PIN'])
df_test['PinDifference'] =abs(df_test['Office_PIN']-df_test['Applicant_City_PIN'])

df_train['Manager_Num_Products_new'] =abs(df_train['Manager_Num_Products']-df_train['Manager_Num_Products2'])
df_test['Manager_Num_Products2_new'] =abs(df_test['Manager_Num_Products']-df_test['Manager_Num_Products2'])
```

```python
# Converting all the date containing columns to datetime datatype and creating new features like applicantAge, managerExperience
# and managerAge from them.

dateColumns = ['Manager_DoB','Manager_DOJ','Applicant_BirthDate','Application_Receipt_Date']
for colName in dateColumns:
    df_train[colName]= pd.to_datetime(df_train[colName])
    df_test[colName]= pd.to_datetime(df_test[colName])
df_train['applicantAge'] = ((df_train['Application_Receipt_Date']-df_train['Applicant_BirthDate'])/ np.timedelta64(1, 'D'))/365
df_train['managerExperience'] = ((df_train['Application_Receipt_Date']-df_train['Manager_DOJ'])/ np.timedelta64(1, 'D'))/365
df_train['managerAge'] =((df_train['Application_Receipt_Date']-df_train['Manager_DoB'])/ np.timedelta64(1, 'D'))/365

df_test['applicantAge'] = ((df_test['Application_Receipt_Date']-df_test['Applicant_BirthDate'])/ np.timedelta64(1, 'D'))/365
df_test['managerExperience'] = ((df_test['Application_Receipt_Date']-df_test['Manager_DOJ'])/ np.timedelta64(1, 'D'))/365
df_test['managerAge'] =((df_test['Application_Receipt_Date']-df_test['Manager_DoB'])/ np.timedelta64(1, 'D'))/365
```

```python
# Creating the feature DateRank which can give the rank values as it specifies the date wise count of the applicant and 
# generating the rank on the basis of their order
prev_date = 0
count_dict = {}
for name, row in df_train.iterrows():
    count_dict[ row["Application_Receipt_Date"] ] = count_dict.get(row["Application_Receipt_Date"],0) + 1

prev_date = 0
rankpct_list = []
for name, row in df_train.iterrows():
    date_value = row["Application_Receipt_Date"]
    if date_value != prev_date:
        rank = 1
        prev_date = date_value
    else:
        rank += 1
    rankpct_list.append(float(rank) / count_dict[date_value])
df_train["DateRank"] = rankpct_list[:]
```

### Class Imbalace Analysis

```python
# Checking for the ratio for both the classes
df_1 = df_train[df_train['Business_Sourced']==1]
df_0 = df_train[df_train['Business_Sourced']==0]

print('No. of datapoints with Business_Sourced val as 1, are :',df_1.shape[0])
print('No. of datapoints with Business_Sourced val as 0, are :',df_0.shape[0])
print('Class Imbalance Ratio :',len(df_0)/len(df_1))
```

**Inference/Approach**
- So, we can see the ratio of Business Not Sourced by the applicant to Business Sourced by the applicant is around 2.
- But in real world scenario, the data actually holds true values, as the Business Sourced count will always be greater than Business Not Sourced So, performing any sampling techniques might lead to in inaccurate results. For e.g. if we go ahead with undersampling, there might be the issue of data constrictions(data loss), whereas if go ahead with oversampling technique, then no. of 'Yes' will be given equal weightage as no. of 'No'. So, this should never be the case.
